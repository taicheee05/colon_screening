```python
import numpy as np
import pandas as pd
from numba import njit, prange
from numba.typed import Dict
from numba.types import int64, float64
from numba.typed import List

# 状態とインデックスの対応を作成
state_list = ['H', 'LR1', 'LR2', 'HR', 'DA', 'DAU', 'DB', 'DBU', 'DC', 'DCU', 'DD', 'DDU', 'D', 'PRH', 'PDAH', 'PDBH', 'PDCH', 'PDDH', 'PRL1', 'PRL2', 'PRLH']
state_map = Dict.empty(key_type=int64, value_type=int64)
for i, state in enumerate(state_list):
    state_map[i] = i


# グローバル変数として定義
change_probability = pd.read_csv('senni_pro.csv').set_index('time').to_numpy()
# 初期状態を設定
v_M_1 = np.array([state_map[0]] * 10000)  # everyone begins in the healthy state
n_i = 10000
n_t = 44
d_c = d_e = 0.02  # コストとQALYsの割引率

# コストの設定
c_ma_f1_dukesA = 1786482
c_ma_f1_dukesB = 2056922
c_ma_f1_dukesC = 2637803
c_ma_f1_dukesD = 3179764
c_ma_f2_dukesA = 34610
c_ma_f2_dukesB = 34610
c_ma_f2_dukesC = 43758
c_ma_f2_dukesD = 2476258
c_fit = 2500
c_tcs = 15000
c_rem_lr_poly = 100000
c_rem_hr_poly = 150000

# ユーティリティの設定
u_dukeA = 0.879
u_dukeB = 0.879
u_dukeC = 0.867
u_dukeD = 0.844
u_normal = 1.0

# 確率の設定
p_res_fit = 0.40
p_speci_fit = 0.946
p_res_tcs_afterFit = 0.40
p_s_fit_lowrisk1 = 0.05
p_s_fit_lowrisk2 = 0.10
p_s_fit_highrisk = 0.15
p_s_fit_dukeA = 0.20
p_s_fit_dukeB = 0.25
p_s_fit_dukeC = 0.30
p_s_fit_dukeD = 0.35
p_tcs_low1 = 0.10
p_tcs_low2 = 0.15
p_tcs_highrisk = 0.20
p_s_tcs_DukeABCD = 0.25
p_perforation_tcs = 0.0001
p_death_after_perforation = 0.05
p_bleeding_with_resection = 0.0001

@njit
def get_substate_probs(current_substate, p_res_fit, p_speci_fit, p_res_tcs_afterFit, p_s_fit_lowrisk1, p_s_fit_lowrisk2, p_s_fit_highrisk, p_s_fit_dukeA, p_s_fit_dukeB, p_s_fit_dukeC, p_s_fit_dukeD, p_tcs_low1, p_tcs_low2, p_tcs_highrisk, p_s_tcs_DukeABCD):
    if current_substate.startswith(('H1', 'LR11', 'LR21', 'HR1', 'DA1', 'DB1', 'DC1', 'DD1', 'PRH1', 'PRL11', 'PRL21', 'PRLH1')):
        return np.array([p_res_fit, 1 - p_res_fit])
    elif current_substate.endswith('11'):
        if current_substate.startswith('H'):
            return np.array([p_speci_fit, 1 - p_speci_fit])
        elif current_substate.startswith('LR1'):
            return np.array([1 - p_s_fit_lowrisk1, p_s_fit_lowrisk1])
        elif current_substate.startswith('LR2'):
            return np.array([1 - p_s_fit_lowrisk2, p_s_fit_lowrisk2])
        elif current_substate.startswith('HR'):
            return np.array([1 - p_s_fit_highrisk, p_s_fit_highrisk])
        elif current_substate.startswith('DA'):
            return np.array([1 - p_s_fit_dukeA, p_s_fit_dukeA])
        elif current_substate.startswith('DB'):
            return np.array([1 - p_s_fit_dukeB, p_s_fit_dukeB])
        elif current_substate.startswith('DC'):
            return np.array([1 - p_s_fit_dukeC, p_s_fit_dukeC])
        elif current_substate.startswith('DD'):
            return np.array([1 - p_s_fit_dukeD, p_s_fit_dukeD])
    elif current_substate.endswith('111'):
        return np.array([p_res_tcs_afterFit, 1 - p_res_tcs_afterFit])
    elif current_substate.endswith('1111'):
        if current_substate.startswith('LR1'):
            return np.array([1 - p_tcs_low1, p_tcs_low1])
        elif current_substate.startswith('LR2'):
            return np.array([1 - p_tcs_low2, p_tcs_low2])
        elif current_substate.startswith('HR'):
            return np.array([1 - p_tcs_highrisk, p_tcs_highrisk])
        elif current_substate.startswith(('DA', 'DB', 'DC', 'DD')):
            return np.array([1 - p_s_tcs_DukeABCD, p_s_tcs_DukeABCD])
    return np.array([1.0])

@njit
def update_substate(current_substate, probs):
    if len(probs) == 1:
        return current_substate + '1'
    else:
        return current_substate + str(np.random.choice(2, p=probs) + 1)

@njit
def get_next_substate(current_state, current_substate, cycle, last_tcs_cycle):
    if current_state == 'H':
        if cycle - last_tcs_cycle <= 5:
            return 'H2'
        elif cycle - last_tcs_cycle > 5:
            return 'H1'
        else:
            return 'H3'
    elif current_state in ['LR1', 'LR2', 'HR', 'DA', 'DB', 'DC', 'DD', 'PRH', 'PRL1', 'PRL2', 'PRLH']:
        if cycle - last_tcs_cycle <= 3:
            return current_state + '2'
        else:
            return current_state + '1'
    else:
        return current_substate
```

```python
@njit
def Probs(M_it, stage, dur, change_probability, state_map, current_substate):
    n_s = len(state_map)
    v_p_it = np.zeros(n_s)
    
    if M_it == state_map[0]:  # 'H'
        if current_substate == 'H2':
            v_p_it[state_map[0]] = 1
        elif current_substate.startswith('H1') or current_substate.startswith('H3'):
            v_p_it[state_map[0]] = 1 - change_probability[stage, 0]
            v_p_it[state_map[1]] = change_probability[stage, 0]
    elif M_it == state_map[1]:  # 'LR1'
        if current_substate == 'LR12':
            v_p_it[state_map[18]] = 1  # PRL1
        elif current_substate in ['LR1112', 'LR11112']:
            v_p_it[state_map[1]] = 1
        elif current_substate == 'LR11111':
            v_p_it[state_map[12]] = p_death_after_perforation * p_perforation_tcs
            v_p_it[state_map[13]] = 1 - p_perforation_tcs
        else:
            v_p_it[state_map[1]] = 1 - change_probability[stage, 1]
            v_p_it[state_map[2]] = change_probability[stage, 1]
    elif M_it == state_map[2]:  # 'LR2'
        if current_substate == 'LR22':
            v_p_it[state_map[19]] = 1  # PRL2
        elif current_substate in ['LR2112', 'LR21112']:
            v_p_it[state_map[2]] = 1
        elif current_substate == 'LR21111':
            v_p_it[state_map[12]] = p_death_after_perforation * p_perforation_tcs
            v_p_it[state_map[13]] = 1 - p_perforation_tcs
        else:
            v_p_it[state_map[2]] = 1 - change_probability[stage, 2]
            v_p_it[state_map[3]] = change_probability[stage, 2]
    elif M_it == state_map[3]:  # 'HR'
        if current_substate == 'HR2':
            v_p_it[state_map[20]] = 1  # PRLH
        elif current_substate in ['HR1112', 'HR11112']:
            v_p_it[state_map[3]] = 1
        elif current_substate == 'HR11111':
            v_p_it[state_map[12]] = p_death_after_perforation * p_perforation_tcs
            v_p_it[state_map[13]] = 1 - p_perforation_tcs
        else:
            v_p_it[state_map[3]] = 1 - change_probability[stage, 3]
            v_p_it[state_map[4]] = change_probability[stage, 3]
    elif M_it == state_map[4]:  # 'DA'
        v_p_it[state_map[4]] = 1 - change_probability[stage, 4] - change_probability[stage, 5]
        v_p_it[state_map[6]] = change_probability[stage, 4]
        v_p_it[state_map[12]] = change_probability[stage, 5]
    elif M_it == state_map[5]:  # 'DAU'
        if dur >= 4:
            v_p_it[state_map[14]] = 1  # 'PDAH'
        else:
            v_p_it[state_map[5]] = 1 - change_probability[stage, 5]
            v_p_it[state_map[12]] = change_probability[stage, 5]
    elif M_it == state_map[6]:  # 'DB'
        v_p_it[state_map[6]] = 1 - change_probability[stage, 6] - change_probability[stage, 7]
        v_p_it[state_map[8]] = change_probability[stage, 6]
        v_p_it[state_map[12]] = change_probability[stage, 7]
    elif M_it == state_map[7]:  # 'DBU'
        if dur >= 4:
            v_p_it[state_map[15]] = 1  # 'PDBH'
        else:
            v_p_it[state_map[7]] = 1 - change_probability[stage, 7]
            v_p_it[state_map[12]] = change_probability[stage, 7]
    elif M_it == state_map[8]:  # 'DC'
        v_p_it[state_map[8]] = 1 - change_probability[stage, 8] - change_probability[stage, 9]
        v_p_it[state_map[10]] = change_probability[stage, 8]
        v_p_it[state_map[12]] = change_probability[stage, 9]
    elif M_it == state_map[9]:  # 'DCU'
        if dur >= 4:
            v_p_it[state_map[16]] = 1  # 'PDCH'
        else:
            v_p_it[state_map[9]] = 1 - change_probability[stage, 9]
            v_p_it[state_map[12]] = change_probability[stage, 9]
    elif M_it == state_map[10]:  # 'DD'
        v_p_it[state_map[10]] = 1 - change_probability[stage, 10]
        v_p_it[state_map[12]] = change_probability[stage, 10]
    elif M_it == state_map[11]:  # 'DDU'
        if dur >= 4:
            v_p_it[state_map[17]] = 1  # 'PDDH'
        else:
            v_p_it[state_map[11]] = 1 - change_probability[stage, 10]
            v_p_it[state_map[12]] = change_probability[stage, 10]
    elif M_it == state_map[12]:  # 'D'
        v_p_it[state_map[12]] = 1
    elif M_it == state_map[13]:  # 'PRH'
        v_p_it[state_map[1]] = change_probability[stage, 11]
        v_p_it[state_map[2]] = change_probability[stage, 12]
        v_p_it[state_map[3]] = change_probability[stage, 13]
        v_p_it[state_map[13]] = 1 - change_probability[stage, 11] - change_probability[stage, 12] - change_probability[stage, 13]
    elif M_it == state_map[14]:  # 'PDAH'
        v_p_it[state_map[4]] = change_probability[stage, 14]
        v_p_it[state_map[14]] = 1 - change_probability[stage, 14]
    elif M_it == state_map[15]:  # 'PDBH'
        v_p_it[state_map[6]] = change_probability[stage, 15]
        v_p_it[state_map[15]] = 1 - change_probability[stage, 15]
    elif M_it == state_map[16]:  # 'PDCH'
        v_p_it[state_map[8]] = change_probability[stage, 16]
        v_p_it[state_map[16]] = 1 - change_probability[stage, 16]
    elif M_it == state_map[17]:  # 'PDDH'
        v_p_it[state_map[10]] = change_probability[stage, 17]
        v_p_it[state_map[17]] = 1 - change_probability[stage, 17]
    elif M_it == state_map[18]:  # 'PRL1'
        v_p_it[state_map[18]] = 1 - change_probability[stage, 1]
        v_p_it[state_map[2]] = change_probability[stage, 1]
    elif M_it == state_map[19]:  # 'PRL2'
        v_p_it[state_map[19]] = 1 - change_probability[stage, 2]
        v_p_it[state_map[3]] = change_probability[stage, 2]
    elif M_it == state_map[20]:  # 'PRLH'
        v_p_it[state_map[4]] = change_probability[stage, 3]
        v_p_it[state_map[20]] = 1 - change_probability[stage, 3]

    assert np.isclose(v_p_it.sum(), 1, atol=1e-6), f"Probabilities do not sum to 1 for state {M_it} and substate {current_substate}"
    return v_p_it
```

```python
@njit
def Costs(M_it, next_M_it, p_it, state_map, current_substate):
    c_it = np.zeros_like(M_it, dtype=np.float64)
    
    for i in range(len(M_it)):
        if current_substate[i].endswith('111'):
            c_it[i] += c_fit
        if current_substate[i].endswith('11121'):
            c_it[i] += c_tcs
        
        if M_it[i] == state_map[4]:  # DA
            if next_M_it[i] == state_map[5] or (next_M_it[i] == state_map[12] and p_it[i, 5] > 0):
                c_it[i] += c_ma_f1_dukesA
        elif M_it[i] == state_map[6]:  # DB
            if next_M_it[i] == state_map[7] or (next_M_it[i] == state_map[12] and p_it[i, 7] > 0):
                c_it[i] += c_ma_f1_dukesB
        elif M_it[i] == state_map[8]:  # DC
            if next_M_it[i] == state_map[9] or (next_M_it[i] == state_map[12] and p_it[i, 9] > 0):
                c_it[i] += c_ma_f1_dukesC
        elif M_it[i] == state_map[10]:  # DD
            if next_M_it[i] == state_map[11] or (next_M_it[i] == state_map[12] and p_it[i, 11] > 0):
                c_it[i] += c_ma_f1_dukesD
        
        if current_substate[i] in ['PRH1111', 'PRL11111', 'PRL21111']:
            c_it[i] += c_rem_lr_poly
        elif current_substate[i] in ['PRLH1111', 'PRLH2111']:
            c_it[i] += c_rem_hr_poly
        
        if M_it[i] == state_map[5]:  # DAU
            c_it[i] += c_ma_f2_dukesA
        elif M_it[i] == state_map[7]:  # DBU
            c_it[i] += c_ma_f2_dukesB
        elif M_it[i] == state_map[9]:  # DCU
            c_it[i] += c_ma_f2_dukesC
        elif M_it[i] == state_map[11]:  # DDU
            c_it[i] += c_ma_f2_dukesD
    
    return c_it
```

```python

@njit
def Effs(M_it, dur, state_map, current_substate, X=None):
    if X is None:
        X = np.ones_like(M_it, dtype=np.float64)
    u_it = np.zeros_like(M_it, dtype=np.float64)
    
    for i in range(len(M_it)):
        if M_it[i] in [state_map[0], state_map[1], state_map[2], state_map[3], state_map[13], state_map[18], state_map[19], state_map[20]]:
            u_it[i] = u_normal
        elif M_it[i] in [state_map[4], state_map[5]] or current_substate[i] in ['DA12', 'DA22']:
            u_it[i] = u_dukeA
        elif M_it[i] in [state_map[6], state_map[7]] or current_substate[i] in ['DB12', 'DB22']:
            u_it[i] = u_dukeB
        elif M_it[i] in [state_map[8], state_map[9]] or current_substate[i] in ['DC12', 'DC22']:
            u_it[i] = u_dukeC
        elif M_it[i] in [state_map[10], state_map[11]] or current_substate[i] in ['DD12', 'DD22']:
            u_it[i] = u_dukeD
        elif M_it[i] == state_map[12]:
            u_it[i] = 0
        elif M_it[i] in [state_map[14], state_map[15], state_map[16], state_map[17]]:
            u_it[i] = u_normal
    
    return u_it * X
```

```python

@njit(parallel=True)
def MicroSim(v_M_1, n_i, n_t, states, change_probability, state_map, 
             p_death_after_perforation, p_perforation_tcs,
             X=None, d_c=0.02, d_e=0.02, TR_out=True, TS_out=True, Trt=False, seed=1):
    np.random.seed(seed)
    v_dwc = 1 / (1 + d_c) ** np.arange(n_t + 1)
    v_dwe = 1 / (1 + d_e) ** np.arange(n_t + 1)
    m_M = np.empty((n_i, n_t + 1), dtype=np.int32)
    m_C = np.zeros((n_i, n_t + 1))
    m_E = np.zeros((n_i, n_t + 1))
    m_M[:, 0] = v_M_1
    dur = np.zeros(n_i, dtype=np.int32)
    last_tcs_cycle = np.full(n_i, -10, dtype=np.int32)
    current_substates = np.full(n_i, 'H1')

    m_C[:, 0] = Costs(m_M[:, 0], m_M[:, 0], np.zeros((n_i, len(state_map))), state_map, current_substates)
    m_E[:, 0] = Effs(m_M[:, 0], dur, state_map, current_substates, X)
    
    for t in range(1, n_t + 1):
        m_p = np.zeros((n_i, len(state_map)))
        for i in prange(n_i):
            m_p[i] = Probs(m_M[i, t-1], t, dur[i], change_probability, state_map, current_substates[i])        
        next_states = samplev(m_p, 1).flatten()
        
        for i in prange(n_i):
            current_substates[i] = get_next_substate(states[next_states[i]], current_substates[i], t, last_tcs_cycle[i])
            substate_probs = get_substate_probs(current_substates[i], p_res_fit, p_speci_fit, p_res_tcs_afterFit)
            current_substates[i] = update_substate(current_substates[i], substate_probs)
            
            if current_substates[i].endswith('11121'):
                last_tcs_cycle[i] = t

        m_C[:, t] = Costs(m_M[:, t - 1], next_states, m_p, state_map, current_substates)
        m_M[:, t] = next_states
        m_E[:, t] = Effs(m_M[:, t], dur, state_map, current_substates, X)
        dur = np.where(np.isin(m_M[:, t], [state_map[5], state_map[7], state_map[9], state_map[11]]), dur + 1, 0)

    tc = m_C @ v_dwc
    te = m_E @ v_dwe
    tc_hat = np.mean(tc)
    te_hat = np.mean(te)
    
    results = {
        "m_M": m_M,
        "m_C": m_C,
        "m_E": m_E,
        "tc": tc,
        "te": te,
        "tc_hat": tc_hat,
        "te_hat": te_hat
    }
    return results
```
